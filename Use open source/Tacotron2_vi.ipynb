<h3>Âm phổ(spectrogram)

Spectrogram (âm phổ)
Nói về âm thanh giống như nhiều sóng, được đặc trưng bởi tần số, bước sóng, chu kỳ, biên độ và vận tốc lan truyền (tốc độ âm thanh).Và âm thanh được biểu thị theo dạng miền thời gian(time domain) với trục y là amplitude (biên độ) và x là thời gian. Để máy tính có thể phân tích, chỉnh sửa và hiểu được âm thanh ta phải trích xuất được dạng sóng(waveform) từ tín hiệu số(được truyển đồi từ âm thanh)

<i>Ví dụ về biểu thị 1 đoạn âm thanh(1 đoạn cá voi hát) xem liệu con người có nghe được không.Mình dung thư viện librosa nổi tiếng của python

```python
import librosa
import librosa.display

filename = 'data/whale_song.wav'
y, sr = librosa.load(filename)
whale_song, _ = librosa.effects.trim(y)
librosa.display.waveplot(whale_song, sr=sr);
plt.ylabel('amptitude');
```

Tín hiệu số bao gồm nhiều sóng âm đơn tần.Khi lấy các mẫu của tín hiệu số theo thời gian thứ mà chúng ta có chỉ là kêt quả của biên độ sóng.Thuật toán The Fourier transform(TFF) cho phép ta tách tín hiệu ra thành các tần số riêng của nó và biên độ của tần số.Tóm cái váy là nó chuyển tín hiệu từ dạng miền thời gian thành miền tín hiệu và kết quả của nó gọi là âm phổ(Spectrum)
<br>Để hiểu rõ hơn các bạn có thể xem <a href="https://www.youtube.com/watch?v=spUNpyF58BY&feature=youtu.be">Video</a> của 3Blue1Brown hoặc đọc <a href="https://en.wikipedia.org/wiki/Fourier_transform">Wikipedia</a>

<b>Dùng thuật toán TFF để convert

```python
import numpy as np
import matplotlib.pyplot as plt

n_fft = 2048
D = np.abs(librosa.stft(whale_song[:n_fft], n_fft=n_fft, hop_length=n_fft+1))
plt.plot(D);
```

Bây giờ ta chia bài hát của cá voi thành các khung thời gian và dùng thuật toán TFF để xem

```python
hop_length = 512
D = np.abs(librosa.stft(whale_song, n_fft=n_fft,  hop_length=hop_length))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='linear');
plt.colorbar();
```

Chả nhìn thấy gì cả âm thanh mà con người có thể nghe được tập trung trong khoảng tần số và biên độ nhỏ

Biến đổ trục y(tần số) sang thang đo logarit(log scale) và trục màu(tay phải - biểu thị biên độ) sang dB.Bạn có thể hiểu là dùng thang đo logarit của biên độ

```python
DB = librosa.amplitude_to_db(D, ref=np.max)
librosa.display.specshow(DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
```

Vùng âm thanh nghe được của tai người nằm trong dải từ 16 - 20.000 Hz (Hertz) hay 16 - 130 dB(đề xi ben)

<b>có thể bạn chưa biết:</b> Sóng hạ âm là sóng âm có tần số thấp hơn 20 Hz(tưởng đương 13dB) , ngưỡng nghe thấy bình thường của người. Khả năng nghe trở nên kém đi khi tần số giảm, vì vậy nên để con người tiếp nhận được hạ âm, áp suất âm thanh cần phải đủ cao. Tai là bộ phận chính có thể cảm nhận sóng hạ âm nhưng ở cường độ cao, những phần khác của cơ thể cũng có thể cảm nhận được sự rung động hạ âm.Một nghiên cứu chỉ ra rằng sóng hạ âm có thể gây ra cảm giác kinh sợ ở người. Có ý kiến cho rằng do không được tiếp nhận một cách có nhận thức, nó có thể khiến con người cảm thấy mơ hồ rằng hiện tượng kì lạ hay siêu nhiên đang diễn ra.Nói tóm gọn là sóng hạ âm chúng ta có thể không nghe thấy nhưng có thể cảm nhận được

<h3>Mel(thang đo Mel)

Khoa học chứng minh rằng con người chúng ta không nhận thức rõ ràng về sự biến đổi tần số trong thang đo tuyến tính.Chúng ta có thể phân biệt sự khác nhau ở tần số thấp, vd sự khác biệt giữa 500 và 1000 Hz rõ ràng trong khi đó 10000 và 10500 Hz rất khó để phân biệt mặc dù khoảng cách là như nhau.<br>
May mắn cho chúng ta là "người nào đó" đã tính hộ hàm biến đổi phi tuyến tính(dùng công thức để biến đổi từ tần số thành thành <i>thang đo mel</i>) và việc ta cần làm là gọi nó ra từ thư viện librosa
